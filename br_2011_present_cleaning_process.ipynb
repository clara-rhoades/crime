## clean legacy zip codes

```python
legacy['ZIP CODE']
```

```python
legacy['ZIP CODE'] = legacy['ZIP CODE'].astype(str)
```

```python
def convert_zip2(value):
    zip_code = ''.join([char for char in value if char.isdigit()])
    return zip_code[:5]
```

```python
# contrib['contbr_zip'] = contrib['contbr_zip'].apply(convert_zip2)
# contrib['contbr_zip'].value_counts(dropna=False)

legacy['ZIP CODE'] = legacy['ZIP CODE'].apply(convert_zip2)
legacy['ZIP CODE'].value_counts(dropna=False)
```
