```python
# function to get value counts on a dataframe
def val_counts(dataframe):
    for col in dataframe.columns:
        print(f'Column name: {col}')
        print(dataframe[col].value_counts(dropna=False))
        print('-' * 50)
        print('')
```

Cleaning tasks
- drop: 'STATION', weather types that aren't 3,8,11,16, all weather type attributes
- change type 'DATE' to datetime
- change col names
    - AWND to AWND = Average daily wind speed (meters per second or miles per hour as per user preference)
    - PGTM to PGTM = Peak gust time (hours and minutes, i.e., HHMM)
    - PRCP = Precipitation (mm or inches as per user preference, inches to hundredths on Daily Form pdf file)
    - TMAX = Maximum temperature (Fahrenheit)
    - TMIN = Minimum temperature (Fahrenheit)
    - WSF2 = Fastest 2-minute wind speed (miles per hour or meters per second as per user preference)
    - WSF5 = Fastest 5-second wind speed (miles per hour or meters per second as per user preference)
    - WT03 = Thunder
        - 1.0 = present; NaN = not present
    - WT08 = Smoke or haze 
    - WT11 = High or damaging winds
    - WT16 = Rain (may include freezing rain, drizzle, and freezing drizzle) 

```python
# look at all the cols when head is called 
pd.set_option('display.max_columns', None)
```

```python
# drop cols
drop_cols = ['STATION','WT01', 'WT02', 'WT04', 'WT05', 'WT06', 'WT13', 'WT17', 'WT19', 'WT21']
for col in weather.columns:
    if 'ATTRIBUTES' in col:
        drop_cols.append(col)

weather.drop(drop_cols,axis=1,inplace=True)
```

```python
weather.head()
```

```python
# rename cols
new_cols = []
for col in weather.columns:
    if col == 'NAME':
        col = 'name'
    if col == 'DATE':
        col = 'date'
    if col == 'AWND':
        col = 'avg_wind_sp'
    if col == 'PGTM':
        col = 'peak_gust'
    if col == 'PRCP':
        col = 'precip'
    if col == 'TMAX':
        col = 'max_temp'
    if col == 'TMIN':
        col = 'min_temp'
    if col == 'WSF2':
        col = '2min_windsp'
    if col == 'WSF5':
        col = '5sec_windsp'
    if col == 'WT03':
        col = 'thunder'
    if col == 'WT08':
        col = 'smoke_or_haze'
    if col == 'WT11':
        col = 'high_or_dmg_winds'
    if col == 'WT16':
        col = 'rain'
    new_cols.append(col)
        
```

```python
# set new col names
weather.columns = new_cols
weather.head()
```
